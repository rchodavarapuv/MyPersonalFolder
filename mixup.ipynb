```python
import numpy as np
grid = np.array([[1,3,5],[2,1,2],[4,3,1]])

a = []

rows, columns = np.shape(grid)
kernel = 2
for i in range(rows-1):
  for j in range(columns-1):
    a.append(np.sum(grid[i:i+kernel,j:j+kernel]))

print(a)
```

```python
import ctypes
a = "babab"

def compared(s):
  temp = s[::-1]
  #temp = strrev(s)
  b = s
  if temp==b:
    print("equal:"+temp)

prefix_arr = ["" for x in range(len(a))]
suffix_arr = ["" for x in range(len(a))]

for i in range(len(a)):
  prefix_arr[i] = a[i:]
  compared(prefix_arr[i])
  suffix_arr[i] = a[:i+1]
  compared(suffix_arr[i])

print(prefix_arr)
print(suffix_arr)

```
